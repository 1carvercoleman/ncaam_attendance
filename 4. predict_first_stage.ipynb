```
import numpy as np
from google.colab import drive
drive.mount('/content/gdrive')
np.random.seed(seed=484)
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

```
np.random.seed(seed=484)
ncaam = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/ncaa_ML_away.csv')
x = ncaam.loc[:, 'people']
z = ncaam.loc[:, 'monthly_cases_pop']
ncaam.loc[:, 'monthly_cases_pop_2'] = [number ** 2 for number in ncaam.loc[:, 'monthly_cases_pop']]
ncaam.loc[:, 'monthly_cases_pop_3'] = [number ** 3 for number in ncaam.loc[:, 'monthly_cases_pop']]
ncaam.loc[:, 'monthly_cases_pop_4'] = [number ** 4 for number in ncaam.loc[:, 'monthly_cases_pop']]
z_2 = ncaam.loc[:, 'monthly_cases_pop_2']
z_3 = ncaam.loc[:, 'monthly_cases_pop_3']
z_4 = ncaam.loc[:, 'monthly_cases_pop_4']
y1 = ncaam.loc[:, 'home_wins']
y2 = ncaam.loc[:, 'point_diff']
y3 = ncaam.loc[:, 'away_turnovers']
X = ncaam.drop(['people', 'attendance', 'game_ids', 'monthly_cases_pop', 'monthly_cases_pop_2', 'monthly_cases_pop_3', 'monthly_cases_pop_4', 'home_wins', 'point_diff', 'home_fg', 'home_three', 'home_ft', 'home_rebound', 'home_assists', 'home_steals', 'home_blocks', 'home_turnovers', 'home_fouls', 'home_t_fouls', 'away_fg', 'away_three', 'away_ft', 'away_rebound', 'away_assists', 'away_steals', 'away_blocks', 'away_turnovers', 'away_fouls', 'away_t_fouls'], axis=1)
from sklearn.preprocessing import StandardScaler
# Create the scaler function
scaler = StandardScaler()
# Fit the function to the X matrix
scaler.fit(X)
# Transform the X matrix and save it as X_scaled
X = scaler.transform(X)
```

```
np.random.seed(seed=484)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

predictors = [x, y1, y2, y3, z, z_2, z_3, z_4]
names = ['people', 'home_wins', 'point_diff', 'away_turnovers', 'monthly_cases_pop', 'monthly_cases_pop_2', 'monthly_cases_pop_3', 'monthly_cases_pop_4']
parameter_grid = {
    'max_depth': [100, 50, 25],
    'min_samples_leaf': [5, 8, 10],
    'n_estimators': [19]}

for i in range(len(predictors)):
  grid_search = GridSearchCV(RandomForestRegressor(),parameter_grid,cv=5,return_train_score=True)
  best_model=grid_search.fit(X,predictors[i])
  hat = best_model.predict(X)
  hat_name = names[i] + "_hat"
  tilda_name = names[i] + "_tilda"
  ncaam.loc[:, hat_name] = hat
  ncaam.loc[:, tilda_name] = ncaam.loc[:, names[i]] - ncaam.loc[:, hat_name]
  print(names[i], " is done")
  print(best_model.best_estimator_.get_params()['max_depth'])
  print(best_model.best_estimator_.get_params()['min_samples_leaf'])

ncaam.to_csv('/content/gdrive/MyDrive/Colab Notebooks/ncaa_ML_away_output.csv')
```
